```python
# Creating a non-numerical season variable
def season_to_string(season):
    if season == 1:
        return 'Winter'
    elif season == 2:
        return 'Spring'
    elif season == 3:
        return 'Summer'
    elif season == 4:
        return 'Fall'
        

hour['season_name'] = hour['season'].map(season_to_string)
display(hour['season_name'].head())
display(hour['season_name'].value_counts())
```

```python
# Small-multiples plot: relplot()
g=sns.relplot(
    data=hour,
    x='hr',
    y='count',
    col='season_name',
    col_wrap=2,
    hue='temp',
    palette='GnBu',
    alpha=0.6,
    height=3
)
g.fig.suptitle('Ridership by Hour', y=1.03);
```

```python
# Regression plot (lmplot)
g=sns.lmplot(
    data=hour,
    x='temp',
    y='count',
    scatter_kws={'color':'lightgrey', 'alpha':0.6}
)
g.fig.suptitle('Ridership vs Temperature', y=1.03);
```

```python
g=sns.lmplot(
    data=hour,
    x='hum',
    y='count',
    scatter_kws={'color':'lightgrey', 'alpha':0.6}
)
g.fig.suptitle('Ridership vs Humidity', y=1.03);
```

```python
# Grouping by date and season for the box plot (avoids plotting too many outliers)
day = hour.groupby(['dteday', 'season_name'])['count'].sum().reset_index()
day.info()
display(day.head())
```

```python
# Categorical plot: Box Plot
g=sns.catplot(
    data=day,
    kind='box',
    x='count',
    y='season_name',
    showmeans=True,
)
g.fig.suptitle('Daily Ridership by Season (Summary)', y=1.03);
```

```python
# Creating a non-numerical weekday variable
def weekday_to_string(weekday):
    if weekday == 6:
        return 'Saturday'
    elif weekday == 0:
        return 'Sunday'
    elif weekday == 1:
        return 'Monday'
    elif weekday == 2:
        return 'Tuesday'
    elif weekday == 3:
        return 'Wednesday'
    elif weekday == 4:
        return 'Thursday'
    elif weekday == 5:
        return 'Friday'
    
hour['weekday_name'] = hour['weekday'].map(weekday_to_string)

display(hour['weekday_name'].head())
display(hour['weekday_name'].value_counts())
```

```python
# Establishing order among weekday values
from pandas.api.types import CategoricalDtype

# Ordering the days
days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

# Creating ordered categorical type
day_type = CategoricalDtype(categories=days, ordered=True)
hour['weekday_name'] = hour['weekday_name'].astype(day_type)
```

```python
# Sanity check for the two newly added columns season_name and weekday_name
hour.head()
```

```python
# The values for the heatmap
display(hour.groupby(['weekday_name', 'mnth'])['count'].mean())

# DataFrame for our heatmap
hour.groupby(['weekday_name', 'mnth'])['count'].mean().unstack()
```

```python
# Heatmap plot
plt.figure(figsize=(10, 7))
sns.heatmap(
    data=hour.groupby(['weekday_name', 'mnth'])['count'].mean().unstack(),
    annot=True,
    fmt='.1f',
    cmap='GnBu'
)
plt.title('Average Ridership Count by Weekday and Month')
plt.xlabel('Month', rotation=.5)
plt.ylabel('Weekday')
plt.yticks(rotation=0)
plt.show()
```

```python
# Sorting the days by mean and median ridership count in descending order
display(hour.groupby('weekday_name')['count'].mean().sort_values(ascending=False))

hour.groupby('weekday_name')['count'].mean().sort_values(ascending=False)
```

```python
# Using melt to prepare our line plot data
melted_hour = hour.melt(
    id_vars=[col for col in hour.columns if col not in ['registered', 'casual']],
    var_name='rider_status',
    value_name='ridership_count'
)
melted_hour
```

```python
# Ridership by registered status DataFrame
melted_day = melted_hour.groupby(['dteday', 'weekday_name', 'rider_status'])['ridership_count'].sum().reset_index()
melted_day
```

```python
# Interactive line plot
import plotly.express as px

line_plot = px.line(
    melted_day,
    x='dteday',
    y='ridership_count',
    color='rider_status',
    line_dash='rider_status',
    title='Registered vs Casual Ridership'
)
line_plot.show()
```

```python
grouped_by_day_melted_day = melted_day.groupby(['weekday_name', 'rider_status'])['ridership_count'].sum().reset_index()
grouped_by_day_melted_day
```

```python
# Interactive bar plot (stacked)
bar_plot = px.bar(
    grouped_by_day_melted_day,
    x='weekday_name',
    y='ridership_count',
    color='rider_status',
    title='Ridership by Weekday'
)
bar_plot.show()
```
